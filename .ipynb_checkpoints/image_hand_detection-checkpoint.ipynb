```python
import mediapipe as mp
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
mp_hands = mp.solutions.hands
mp_drawing = mp.solutions.drawing_utils
mesh_drawing_spec = mp_drawing.DrawingSpec(thickness=2, color=(0, 255, 0))
mark_drawing_spec = mp_drawing.DrawingSpec(thickness=3, circle_radius=3, color=(0, 0, 255))
```

```python
img_path = "hand3.jpg"
```

```python
plt.imshow(plt.imread(img_path))
```

## mediapipeによるての検出 

```python
with mp_hands.Hands(
        max_num_hands=2,
        min_detection_confidence=0,
        static_image_mode=True) as hands_detection:
    
    image = cv2.imread(img_path)
    # image = cv2.resize(image, dsize=None, fx=0.3, fy=0.3)
    rgb_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    height = rgb_image.shape[0]
    width = rgb_image.shape[1]
    
    results = hands_detection.process(rgb_image)
    
    annotated_image = image.copy()
```

## 検出したてにランドマークを描画

```python
for hand_landmarks in results.multi_hand_landmarks:
    # for id, lm in enumerate(hand_landmarks.landmark):
    #     print(id, lm.x)
    mp_drawing.draw_landmarks(
        image=annotated_image,
        landmark_list=hand_landmarks,
        connections=mp_hands.HAND_CONNECTIONS,
        landmark_drawing_spec=mark_drawing_spec,
        connection_drawing_spec=mesh_drawing_spec
        )
    # cv2.imwrite("result.jpg", annotated_image)
    #annotated_image = cv2.cvtColor(annotated_image, cv2.COLOR_BGR2RGB)
    plt.imshow(annotated_image)
```

## 検出したての人差し指と親指を直線で結ぶ

```python
# 元画像のサイズ
annotated_image_h, annotated_image_w, _d = annotated_image.shape
print(f"元画像の高さ： {annotated_image_h}")
print(f"元画像の幅： {annotated_image_w}")
```

```python
# 座標を画像の解像度に変換
x = int(results.multi_hand_landmarks[0].landmark[8].x * annotated_image.shape[0])
y = int(results.multi_hand_landmarks[0].landmark[8].y * annotated_image.shape[1])
print(x)
print(y)
```

```python
cv2.putText(annotated_image,
            text='finger',
            org=(x, y),
            fontFace=cv2.FONT_HERSHEY_SIMPLEX,
            fontScale=1.0,
            color=(0, 255, 0),
            thickness=2,
            lineType=cv2.LINE_4)
rgb_annotated_image = cv2.cvtColor(annotated_image, cv2.COLOR_BGR2RGB)
plt.imshow(annotated_image)
```

```python
# 人差し指先端の座標（０〜１）
index_finger = np.array([results.multi_hand_landmarks[0].landmark[8].x, results.multi_hand_landmarks[0].landmark[8].y])
# 親差し指先端の座標（０〜１）
index_thumb = np.array([results.multi_hand_landmarks[0].landmark[4].x, results.multi_hand_landmarks[0].landmark[4].y])
print(index_finger)
print(index_thumb)
```

```python
_x = index_finger * annotated_image.shape[1]
x = _x.astype(int)
_y = index_thumb * annotated_image.shape[0]
y = _y.astype(int)
print(x)
print(y)
```

```python
cv2.line(annotated_image, pt1=x, pt2=y, color=(0, 0, 255), thickness=2)
rgb_annotated_image = cv2.cvtColor(annotated_image, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_annotated_image)
```

## 検出したてにランドマークを描画

```python
for hand_landmarks in results.multi_hand_landmarks:
    # for id, lm in enumerate(hand_landmarks.landmark):
    #     print(id, lm.x)
    mp_drawing.draw_landmarks(
        image=annotated_image,
        landmark_list=hand_landmarks,
        connections=mp_hands.HAND_CONNECTIONS,
        landmark_drawing_spec=mark_drawing_spec,
        connection_drawing_spec=mesh_drawing_spec
        )
    cv2.imwrite("result.jpg", annotated_image)
```
